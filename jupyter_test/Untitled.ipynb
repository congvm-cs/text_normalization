```python
def get_IPA(IPA_CMU):
    #convert CMU to standard IPA
    IPA = "" # final string
    IPA_CMU = IPA_CMU.split()
    for w in IPA_CMU:
        if w == "a":   w = w.replace("a", "ə")
        if "ey" in w:  w = w.replace("ey", "e")
        if "aa" in w:  w = w.replace("aa", "ɑ")    
        if "ae" in w:  w = w.replace("ae", "æ")
        if "ah" in w:  w = w.replace("ah", "ə")
        if "ao" in w:  w = w.replace("ao", "ɔ")
        if "aw" in w:  w = w.replace("aw", "aʊ")
        if "ay" in w:  w = w.replace("ay", "aɪ")
        if "ch" in w:  w = w.replace("ch", "ʧ")
        if "dh" in w:  w = w.replace("dh", "ð")
        if "eh" in w:  w = w.replace("eh", "ɛ")
        if "er" in w:  w = w.replace("er", "ər")
        if "hh" in w:  w = w.replace("hh", "h")
        if "ih" in w:  w = w.replace("ih", "ɪ")
        if "jh" in w:  w = w.replace("jh", "ʤ")
        if "ng" in w:  w = w.replace("ng", "ŋ")
        if "ow" in w:  w = w.replace("ow", "oʊ")
        if "oy" in w:  w = w.replace("oy", "ɔɪ")
        if "sh" in w:  w = w.replace("sh", "ʃ")
        if "th" in w:  w = w.replace("th", "θ")
        if "uh" in w:  w = w.replace("uh", "ʊ")
        if "uw" in w:  w = w.replace("uw", "u")
        if "zh" in w:  w = w.replace("zh", "ʒ")
        if "iy" in w:  w = w.replace("iy", "i")
        if "y" in w:   w = w.replace("y", "j")
        IPA += w + " "
    return(IPA.strip());
```

```python
import re
import os
import json
from eng_to_ipa import transcribe
import numpy as np
import unicodedata

with open('./phones.json', "r") as phones_json:
    PHONES = json.load(phones_json)

# list of adjacent vowel symbols that constitute separate nuclei
hiatus = [["er", "iy"], ["iy", "ow"], ["uw", "ow"], ["iy", "ah"], ["iy", "ey"], ["uw", "eh"], ["er", "eh"]]
consonant_list = ['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']
vowel_list = ['a', 'ă', 'â', 'e', 'ê', 'i', 'o', 'ô', 'ơ', 'u', 'ư', 'y']

def cmu_syllable_count(word):
    """count syllables based on CMU transcription"""
    word = re.sub("\d", "", word).split(' ')
    if "__IGNORE__" in word[0]:
        return 0
    else:
        nuclei = 0
        index = []
        for i, sym in enumerate(word):
            prev_phone = PHONES[word[i-1]]
            prev_sym = word[i-1]
            if PHONES[sym] == 'vowel':
                if i > 0 and not prev_phone == 'vowel' or i == 0:
                    nuclei += 1
                    index.append(i)
                elif [prev_sym, sym] in hiatus:
                    nuclei += 1
                    index.append(i)
        return nuclei, index


def syllable_count(word: str):
    """transcribes a regular word to CMU to fetch syllable count"""
    if len(word.split()) > 1:
        return [syllable_count(w) for w in word.split()]
    word = transcribe.get_cmu([transcribe.preprocess(word)])
    return cmu_syllable_count(word[0][0])


def split_in_pronunciation(word):
    text = word.lower()
    try:
        text = [t.lower() for t in cmu_dict[text][0]]
        print(text)
        text = [get_IPA(i) for i in text]
        print(text)
        text = [get_viet_sound(i) for i in text]    
        text.extend(['_', '_'])
        print(text)

        c, stop_nodes = syllable_count(word)
        print(stop_nodes)
        ret = []

        for i in stop_nodes:    
            # babba -> bab ba
            # baba -> ba ba
            if (text[i+1] in consonant_list) and (text[i+2] not in consonant_list):
                ret.append(text[i-1:i+1])

            elif (text[i+1] in consonant_list) and (text[i+2] in consonant_list):
                ret.append(text[i-1:i+2])
            try:
                for i in ret:
                    i.remove('_')
            except:
                pass
    except:
        pass
    
    return text, ret
```
